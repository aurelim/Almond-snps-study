# Fastsimcoal2 simulated populations

```python
import os
import allel
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dossier contenant les fichiers de g√©notypes
input_folders_pop = "/shared/projects/almond_dw_snp/4_FastSimCoal/results/fastsimcoal2/simulations_2"
```

```python
data = pd.read_csv(os.path.join(input_folders_pop, 'pi_values_in_fsc_sim_pops.csv'), sep=',', names=['Population', 'Chromosome', 'Pi'])

import seaborn as sns


couleurs = {
    'Pdulcis_C_Asia': '#8d5bb3',
    'Pdulcis_Euro_1': '#47c5f3',
    'Pdulcis_Euro_2': '#ffbb86',
    'Pdulcis_N_Amer': '#cb3636',
    'Pfenzliana': '#ffe38b',
    'Porientalis': '#283f75',
    'Pspinosissima': '#7b9e64'
}

# Cr√©ation des boxplots pour œÄ, Ho et He
plt.figure(figsize=(8, 15))

# Boxplot pour œÄ
plt.subplot(3, 1, 1)
sns.boxplot(x='Chromosome', y='Pi', hue='Population', palette=couleurs, data=data, fliersize=1)
plt.title('Diversity œÄ by Population and Chromosome')
plt.ylabel('œÄ (Diversity)')
plt.xticks(rotation=45)
plt.ylim(0,0.01)
plt.legend(title='Almond populations', loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=4)  # Ajuste selon tes besoins

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Exemple de donn√©es (remplace cela par pi_values_all_chromosomes, He_values_all_chromosomes, Ho_values_all_chromosomes)
# pi_values_all_chromosomes = {
#     '1': [0.1, 0.2, 0.3, ...],
#     '2': [0.4, 0.5, 0.6, ...],
#     ...
# }
# He_values_all_chromosomes = {...}
# Ho_values_all_chromosomes = {...}

# Cr√©er une figure avec 3 sous-graphes
fig, axs = plt.subplots(3, 1, figsize=(6, 12))

# Distribution des valeurs de diversit√© œÄ
axs[0].boxplot(
    [pi_values_all_chromosomes[chrom] for chrom in sorted(pi_values_all_chromosomes.keys())],
    labels=[f"Chromosome {chrom}" for chrom in sorted(pi_values_all_chromosomes.keys())],
    showfliers=False
)
for i, chrom in enumerate(sorted(pi_values_all_chromosomes.keys())):
    pi_values = pi_values_all_chromosomes[chrom]
    jittered_x = np.random.normal(i + 1, 0.1, size=len(pi_values))
    axs[0].scatter(jittered_x, pi_values, alpha=0.5, color='red', s=5)
axs[0].set_title("Distribution des valeurs de diversit√© œÄ par chromosome (tous r√©plicats confondus)")
axs[0].set_xlabel("Chromosome")
axs[0].set_ylabel("Diversit√© œÄ")
axs[0].tick_params(axis='x', rotation=45)

# Distribution des valeurs de He
axs[1].boxplot(
    [He_values_all_chromosomes[chrom] for chrom in sorted(He_values_all_chromosomes.keys())],
    labels=[f"Chromosome {chrom}" for chrom in sorted(He_values_all_chromosomes.keys())],
    showfliers=False
)
for i, chrom in enumerate(sorted(He_values_all_chromosomes.keys())):
    he_values = He_values_all_chromosomes[chrom]
    jittered_x = np.random.normal(i + 1, 0.1, size=len(he_values))
    axs[1].scatter(jittered_x, he_values, alpha=0.5, color='blue', s=5)
axs[1].set_title("Distribution des valeurs de He par chromosome (tous r√©plicats confondus)")
axs[1].set_xlabel("Chromosome")
axs[1].set_ylabel("Expected heterozygosity")
axs[1].tick_params(axis='x', rotation=45)

# Distribution des valeurs de Ho
axs[2].boxplot(
    [Ho_values_all_chromosomes[chrom] for chrom in sorted(Ho_values_all_chromosomes.keys())],
    labels=[f"Chromosome {chrom}" for chrom in sorted(Ho_values_all_chromosomes.keys())],
    showfliers=False
)
for i, chrom in enumerate(sorted(Ho_values_all_chromosomes.keys())):
    ho_values = Ho_values_all_chromosomes[chrom]
    jittered_x = np.random.normal(i + 1, 0.1, size=len(ho_values))
    axs[2].scatter(jittered_x, ho_values, alpha=0.5, color='green', s=5)
axs[2].set_title("Distribution des valeurs de Ho par chromosome (tous r√©plicats confondus)")
axs[2].set_xlabel("Chromosome")
axs[2].set_ylabel("Observed heterozygosity")
axs[2].tick_params(axis='x', rotation=45)

# Ajuster l'espacement entre les sous-graphes
plt.tight_layout()
plt.show()
```
